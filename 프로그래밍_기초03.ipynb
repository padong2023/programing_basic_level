# if문 심화 - elif(else if)
여러가지 조건을 넣고 싶을 때 사용(ex. mbti)




````python
print("좋아하는 음식 선택하기")
print("1. 마라탕, 2. 탕후루, 3. 국밥, 4. 삼겹살")
a = int(input("숫자를 입력해주세요: "))

if a==1:
  print("너도 마라탕 좋아하는구나~! 나도 좋아해")
elif a==2:
  print("나는 탕후루 안좋아해ㅠㅠ")
elif a==3:
  print("너 국밥충이야?")
elif a==4:
  print("삽겹살에 별빛청하ㄱㄱ??")
else :    # 실패 메시지
  print("잘못입력했습니다.")
````

# 제어문 - while문



````python
# 문제 상황2-1 : 10번 찍어 안 넘어가는 나무 없다.
treehit = 1

while treehit<11:
  print("나무를 %d번 찍었습니다." %treehit)
  if treehit==10:
    print("나무가 넘어갑니다.")
  treehit = treehit + 1     # treehit += 1 과 같은 말
````

# 제어문 - for문


```
for 변수 in 리스트(또는 튜플, 문자열) :
  수행할 문장1
  수행할 문장2
```

````python
# 전형적인 for문 예시
test_list = ['one','two', 'three']
for i in test_list:
  print(i)
````

````python
# 다양한 for문의 사용
a = [(1,2),(3,4), (5,6)]
for (h, l) in a:
  u = h+l
  print(u)
````

````python
# for문의 응용
marks = [90, 25, 67, 45, 80]

num = 1
for mark in marks:
  if mark >= 60:
    print("%d번 학생은 합격입니다." %num)
  else:
    print("%d번 학생은 불합격입니다." %num)
  num = num + 1
````

````python
# for문과 continue
marks = [90, 25, 67, 45, 80]

num = 0
for mark in marks:
  num = num + 1
  if mark < 60:
    continue
  print("%d번 학생은 합격입니다." %num)
````

````python
# for문과 함께 자주 사용하는 range함수
# ex) range(0,10)-> 0이상 10미만
marks = [90, 25, 67, 45, 80]

for num in range(len(marks)) :      #len은 길이를 숫자로 나타내주는 함수
  if(marks[num]<60):
    continue
  print("%d번 학생 축하합니다. 합격입니다." %(num+1))
````

````python
# 중첩 for문과 range를 이용한 구구단
for i in range(2, 20):
  print("%d단 "%i)
  for j in range(1, 10):
    print(i*j, end ="|")
  print('')
  print('')
````

````python
# 리스트 안에 for문 포함하기
a = [1, 2, 3, 4]
result = []

for num in a:
  result.append(num*3)    #append는 추가하기 기능

print(result)
````

````python
# 위의 코드 변형v.1
a = [1, 2, 3, 4]
result = [num*3 for num in a]
print(result)
````

# 리스트 안에 조건문을 포함할 수 있다~!


```
[표현식 for 항목 in 반복 가능 객체 if 조건]
```

# for문을 여러 개 사용하는 방법


```
[표현식 for 항목1 in 반복 가능 객체1 if 조건1]
  for 항목2 in 반복 가능 객체2 if 조건2
  for 항목3 in 반복 가능 객체3 if 조건3
  ....
  for 항목n in 반복 가능 객체n if 조건n
```




````python
# 위의 코드 변형v.2
a = [1, 2, 3, 4]
result = [num*3 for num in a if num%2==0]
print(result)
````

````python
# 구구단 변형
result = [x*y for x in range(2,10)
                for y in range(1,10)]

print(result)
````

````python
## mbti 검사 코드 만들기

print("MBTI 검사를 시작합니다. 다음 다섯 가지 질문에 대한 답변을 선택해주세요.")

# 다섯 가지 질문
questions = [
    "1. 새로운 사람들과 쉽게 친구가 됩니다. (a) 그렇다 (b) 아니다",
    "2. 대부분의 상황에서 조용하고 내성적입니다. (a) 그렇다 (b) 아니다",
    "3. 계획을 세우는 것을 좋아합니다. (a) 그렇다 (b) 아니다",
    "4. 무엇이든지 시도하고 경험하는 것을 좋아합니다. (a) 그렇다 (b) 아니다",
    "5. 감정적이며 다른 사람들의 감정에 민감합니다. (a) 그렇다 (b) 아니다"
]

# 사용자의 답변을 저장할 변수
answers = []

# 다섯 가지 질문에 대한 답변을 입력 받습니다.
for question in questions:
    answer = input(question + " (a 또는 b로 답해주세요): ")
    while answer not in ['a', 'b']:
        print("잘못된 입력입니다. 'a' 또는 'b' 중에서 선택해주세요.")
        answer = input(question + " (a 또는 b로 답해주세요): ")
    answers.append(answer)

# 답변에 따라 MBTI 유형을 결정합니다. - mbti 알고리즘
mbti_type = ""
if answers[0] == 'a':
    if answers[1] == 'a':
        mbti_type += 'E'      # 첫번째 질문과 두번째 질문의 답변이 "a"인 경우
    else:
        mbti_type += 'I'      # 첫번째 질문"a", 두번째 질문의 답변이 "b"인 경우
else:
    if answers[1] == 'a':
        mbti_type += 'I'      # 첫번째 질문"b", 두번째 질문의 답변이 "a"인 경우
    else:
        mbti_type += 'E'      # 첫번째 질문"b", 두번째 질문의 답변이 "b"인 경우

if answers[2] == 'a':
    mbti_type += 'N'
else:
    mbti_type += 'S'

if answers[3] == 'a':
    mbti_type += 'T'
else:
    mbti_type += 'F'

if answers[4] == 'a':
    mbti_type += 'P'
else:
    mbti_type += 'J'

# MBTI 유형을 출력합니다.
print("당신의 MBTI 유형은:", mbti_type)
````
